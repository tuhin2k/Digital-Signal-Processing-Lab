```
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(25,25))
i=1

a=1
lim=1
nT=np.linspace(-lim,lim,161)
x_nT=a*np.cos(10*np.pi*nT)

def DrawGraph (x,y) :
  global i
  plt.subplot(6,1,i)
  plt.stem(x,y)
  plt.axvline(linewidth='1',color='r')
  plt.axhline(linewidth='1',color='r')
  i=i+1


DrawGraph(nT,x_nT)

def Delay(x_n,k):
  ret=np.zeros(abs(k))
  if(k>=0):
    ret=np.concatenate((ret,x_n[0:len(x_n)-k]))
  else:
    ret=np.concatenate((x_n[abs(k):len(x_n)],ret))

  return ret

DrawGraph(nT,Delay(x_nT,5))

DrawGraph(nT,Delay(x_nT,-8))

x_nT_4=nT*0
x_nT_4[80]=1

DrawGraph(nT,Delay(x_nT_4,-3))

aa=np.zeros(81)
aa[80]=1
bb=np.zeros(80)
for j in range(0,80) :
  bb[j]=1
x_nT_5=np.concatenate((aa,bb))

DrawGraph(nT,Delay(x_nT_5,-2))

for j in range(0,80) :
  bb[j]=j+2
x_nT_6=np.concatenate((aa,bb))

DrawGraph(nT,x_nT_6)

```
