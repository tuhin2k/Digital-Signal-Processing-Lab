```python
import matplotlib.pyplot as plt
import numpy as np

# Computing and plotting Discrete Fourier Transform (DFT) of a signal using DFT equation.

def DFT(x):
    """
    Function to calculate the 
    discrete Fourier Transform 
    of a 1D real-valued signal x
    """

    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    
    X = np.dot(e, x)
    
    return X


sr = 100 # sampling rate

ts = 1.0/sr # sampling interval
t = np.arange(0,1,ts)

freq = 1.
x = 3*np.sin(2*np.pi*freq*t)

freq = 5
x += np.sin(2*np.pi*freq*t)

freq = 9   
x += 0.5* np.sin(2*np.pi*freq*t)

plt.figure(figsize = (18, 15))
plt.subplot(2,1,1)
plt.plot(t, x, 'g')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')

plt.subplot(2,1,2)

X = DFT(x)

# calculate the frequency
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 


plt.stem(freq, abs(X),'b')
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')

plt.subplots_adjust(hspace=.5)
```

```python
# Computing the inverse DFT of the signal obtained in the previous program .
def IDFT(x):
  N = len(x)
  n = np.arange(N)
  k = n.reshape((N,1))
  e = np.exp(1j*2*np.pi*k*n/N)
  e/=N
  res = np.dot(e,x)
  return res


signal_x = IDFT(X) # X is recieved from the previous DFT program

ts = 1.0/100
t = np.arange(0,1,ts)

plt.figure(figsize = (18, 7))
plt.title('We are getting the same signal after computing the inverse DFT as like as the input signal of the previous DFT program')
plt.plot(t,signal_x)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
```
