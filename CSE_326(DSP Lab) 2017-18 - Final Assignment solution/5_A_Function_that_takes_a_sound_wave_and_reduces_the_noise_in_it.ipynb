We can use noise reduce library of pypi to reduce background noise from a audio signal.

Noise reduction in python:

The algorithm requires two inputs:A signal audio clip containing the signal and the noise intended to be removed and datarate of this signal.

Steps of algorithm-

1.An FFT is calculated over the noise audio clip.

2.Statistics are calculated over FFT of the the noise (in frequency).

3.A threshold is calculated based upon the statistics of the noise (and the desired sensitivity of the algorithm).

4.An FFT is calculated over the signal.

5.A mask is determined by comparing the signal FFT to the threshold.

6.The mask is smoothed with a filter over frequency and time.

7.The mask is appled to the FFT of the signal, and is inverted.

```python
import noisereduce as nr
import numpy as np 
import matplotlib.pyplot as plt 
import wave
from IPython.display import Audio
from scipy.io import wavfile

def noise_reductor(audiofile):
  #load data 
  rate, data = wavfile.read("noisyspeech.wav") 
  # perform noise reduction 
  reduced_noise = nr.reduce_noise(data,rate)

  return data,reduced_noise,rate


inputaudio,noise_reduced_signal,sampling_rate= noise_reductor('noisyspeech.wav')
plt.figure(figsize=(25,15))

plt.subplot(2,1,1)
plt.plot(inputaudio)
plt.title("The input audio clip signal with noises")
plt.subplot(2,1,2)
plt.plot(noise_reduced_signal)
plt.title("After reducing noises,The audio clip signal")



```
