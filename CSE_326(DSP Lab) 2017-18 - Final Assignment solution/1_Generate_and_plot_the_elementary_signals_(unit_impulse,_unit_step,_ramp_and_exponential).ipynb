```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi



plt.figure(figsize=(20,18))

n= np.arange(-10,11,1)
l=np.size(n)
impulse=np.zeros(l)
for i,j in enumerate(n):
  # print("i= " ,i ,"j= ",j)
  if j==0 :
    impulse[i]=1
    break
plt.subplot(4,1,1)
plt.stem(n,impulse)
plt.title('Unit Impulse Signal')
plt.xlabel('Sample'); plt.ylabel('Amplitude')


stp=np.zeros(l)
for i,j in enumerate(n):
  # print("i= " ,i ,"j= ",j)
  if j>=0 :
    stp[i]=1

plt.subplot(4,1,2)
plt.stem(n,stp)
plt.title('Unit Step Signal')
plt.xlabel('Sample'); plt.ylabel('Amplitude')


ramp=np.zeros(l)
for i,j in enumerate(n):
  # if j==0 :
  #   continue
  
  if j>0:
    ramp[i]=1 + ramp[i-1]

  # print("i= " ,i ,"j= ",j)
  

plt.subplot(4,1,3)
plt.stem(n,ramp)
plt.title('Ramp Signal')
plt.xlabel('Sample'); plt.ylabel('Amplitude')


a=4
b=5
x=np.arange(0,3.14,0.01)
exp=np.exp(a*x+b)


plt.subplot(4,1,4)
plt.plot(x,exp)
plt.title('Exponential Signal')
plt.xlabel('Sample'); plt.ylabel('Amplitude')



plt.subplots_adjust(hspace=1)
```
