```
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(25,20))
lim=1.0
#analog signal
a=1.0
f=5.0
d= -np.pi/2.0
t= np.linspace(-lim,lim,1000)
x_t=a*np.cos(2.0*np.pi*f*t + d)
plt.subplot(3,1,1)
plt.plot(t,x_t)
#plt.stem(t,x_t)
#sampling
Fs= 150.0
T=1.0 /Fs
nT= np.arange(-lim,lim,T)
x_nT=a*np.cos(2.0*np.pi*f*nT + d)
plt.subplot(3,1,2)
#plt.plot(nT,x_nT)
plt.stem(nT,x_nT)
#quantizing
bits=2
levels= 2 ** (bits-1)
step=1.0/levels
xq_n=np.round(x_nT/step)*step
plt.subplot(3 ,1, 3)
plt.stem(nT,xq_n)
```
